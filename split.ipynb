```python
import os
import shutil
import random
from tqdm import tqdm

SRC_DIR = "vessel12_dataset" # original dataset
OUT_DIR = "vessel12_split"   # output

IMG_DIR = os.path.join(SRC_DIR, "vessel12_images")
MSK_DIR = os.path.join(SRC_DIR, "VESSEL12_01-20_Lungmasks")

# create output dirs
for split in ["train_val", "test"]:
    os.makedirs(os.path.join(OUT_DIR, split, "images"), exist_ok=True)
    os.makedirs(os.path.join(OUT_DIR, split, "masks"), exist_ok=True)

print("Directories ready.")
```

```python
files = sorted([f for f in os.listdir(IMG_DIR) if f.endswith(".mhd")])

print("Total scans:", len(files))

random.seed(42)
random.shuffle(files)
```

```python
n = len(files)

train_val_split = int(0.8 * n)

train_val_files = files[:train_val_split]
test_files      = files[train_val_split:]

print(f"Train+Val: {len(train_val_files)}, Test: {len(test_files)}")
```

```python
def copy_pair(filename, split):
    base = filename.replace(".mhd", "")

    # image paths
    img_mhd = os.path.join(IMG_DIR, base + ".mhd")
    img_raw = os.path.join(IMG_DIR, base + ".raw")

    # mask paths (adjust if needed)
    msk_mhd = os.path.join(MSK_DIR, base + ".mhd")
    msk_raw = os.path.join(MSK_DIR, base + ".raw")

    dst_img = os.path.join(OUT_DIR, split, "images")
    dst_msk = os.path.join(OUT_DIR, split, "masks")

    # fail fast
    for f in [img_mhd, img_raw, msk_mhd, msk_raw]:
        if not os.path.exists(f):
            raise FileNotFoundError(f"Missing: {f}")

    shutil.copy(img_mhd, dst_img)
    shutil.copy(img_raw, dst_img)

    shutil.copy(msk_mhd, dst_msk)
    shutil.copy(msk_raw, dst_msk)
```

```python
for f in tqdm(train_val_files):
    copy_pair(f, "train_val")

for f in tqdm(test_files):
    copy_pair(f, "test")

print("Split complete.")
```

```python
sample = train_val_files[0].replace(".mhd", "")

img_path = os.path.join(OUT_DIR, "train_val", "images", sample + ".mhd")
raw_path = os.path.join(OUT_DIR, "train_val", "images", sample + ".raw")

print("Image exists:", os.path.exists(img_path))
print("Raw exists:", os.path.exists(raw_path))
```
